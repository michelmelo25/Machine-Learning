```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("wdbc.data", header=-1)
df.head()
```

#### Removendo a primeira coluna que é o ID

```python
df.drop(columns=[0], axis=1, inplace=True)
df.head()
```

```python
df = pd.get_dummies(df, drop_first=True)
```

```python
df.head()
```

# Separando os dados

```python
X = df.values[: , :-1]
y = df.values[:, -1]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=42)
# X_test, X_val, y_test, y_val = train_test_split(X_test, y_test, test_size=.5, random_state=42)
```

```python
X_train.shape, y_train.shape, X_test.shape,  y_test.shape #, X_val.shape,   y_val.shape
```

# Grid Search

```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV, StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
```

```python
learning_rate = [0.1, 0.05, 0.01]
n_estimators = [50, 100, 200]
max_depth = [3, 5, 7]

pipe = {}
pipe['dt'] = Pipeline([('transformer', StandardScaler()),('dt', DecisionTreeClassifier())])
pipe['rf'] = Pipeline([('transformer', StandardScaler()),('rf', RandomForestClassifier())])
pipe['gb'] = Pipeline([('transformer', StandardScaler()),('gb', GradientBoostingClassifier())])

params = {}
params['dt'] = {'dt__max_depth': max_depth}
params['rf'] = {'rf__n_estimators': n_estimators, 'rf__max_depth': max_depth}
params['gb'] = {'gb__learning_rate': learning_rate, 'gb__n_estimators': n_estimators, 'gb__max_depth': max_depth}

cv = StratifiedKFold(n_splits=4, shuffle=True, random_state=42)
gs_result = {}
for key in pipe:
    gs = GridSearchCV(pipe[key], params[key], cv=cv, iid=True, scoring='accuracy', return_train_score=True)
    gs.fit(X_train, y_train)
    gs_result[key] = gs.cv_results_
```

# DecisionTreeClassifier

```python
cols = ['params', 'mean_train_score', 'mean_test_score']
pd.DataFrame(gs_result['dt'])[cols].sort_values(by='mean_test_score', ascending=False)
```

```python
dt = DecisionTreeClassifier(max_depth=5)
dt.fit(X_train, y_train)
dt_pred = dt.predict(X_test)
accuracy_score(dt_pred, y_test)
```

### Foi obtido uma melhor acurácia usando o dataset inteiro
## Importância das features

```python
dt.feature_importances_ # muitas features têm importância 0
```

# RandomForestClassifier

```python
cols = ['params', 'mean_train_score', 'mean_test_score']
pd.DataFrame(gs_result['rf'])[cols].sort_values(by='mean_test_score', ascending=False)
```

```python
rf = RandomForestClassifier(max_depth=7, n_estimators=100)
rf.fit(X_train, y_train)
rf_pred = rf.predict(X_test)
accuracy_score(rf_pred, y_test)
```

### Foi obtido uma melhor acurácia usando o dataset inteiro
## Importância das features

```python
rf.feature_importances_ # Todos os atributos têm certa importância
```
